```python
#Lowercasing
df2['overview'] = df2['overview'].str.lower()
```

```python
#Remove punctuations
import string
exclude = string.punctuation
def remove_punctuation(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
df2['overview'] = df2['overview'].apply(remove_punctuation)
```

```python
df3 = df2.copy()
```

```python
from textblob import TextBlob

def remove_spelling_mistakes(text):
  textblob = TextBlob(text)
  textblob.correct().string

df3['overview'].apply(remove_spelling_mistakes)
```

```python
from textblob import TextBlob

def batch_correct_text(text_list):
    text_blob = TextBlob(" ".join(text_list))
    corrected = text_blob.correct()
    return corrected.string.split()

batch_size = 100  # Experiment with different batch sizes
for i in range(0, len(df3), batch_size):
    texts = df3['overview'].iloc[i:i+batch_size].tolist()
    corrected_texts = batch_correct_text(texts)
    df3['overview'].iloc[i:i+batch_size] = corrected_texts
```

```python
#Removing stop words

from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

def remove_stopwords(text):
  new_text = []
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)

  x=new_text[:]
  new_text.clear
  return ' '.join(x)

remove_stopwords('Hi, my name is ashish and I am from the beautiful city of Kanpur')
```

```python
#Tokenization
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize

def get_tokenize(text):
    return word_tokenize(text)

tokens = get_tokenize('Hi, I am ashish malhotra born and brought up in Kanpur')
print(tokens)
```

```python
#Stemming

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem_word(text):
  return ' '.join([ps.stem(word) for word in text.split()])

stem_word("walk walked walking")
```
