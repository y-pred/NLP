Text Preprocessing

```python
#For Recommendation part we will take original dataframe
df.head()
```

```python
df3 = df[['title','director','cast','country','rating','listed_in','description','type']]
```

```python
df3.fillna('',inplace  = True)
```

```python
df3.sample(5)
```

```python
import string
exclude = string.punctuation

def preprocess(text):
  temp = text.split(', ')
  temp_arr = []
  for i in temp:
    temp_text = str.lower(i).translate(str.maketrans('','',exclude))
    temp_arr.append(temp_text.replace(' ',''))

  return(' '.join(temp_arr))

preprocess('Emilio- Palacios, Jorge !$Clemente, Beatriz Medin')
```

```python
df3['type'] = df3['type'].apply(preprocess)
df3['director'] = df3['director'].apply(preprocess)
df3['cast'] = df3['cast'].apply(preprocess)
df3['country'] = df3['country'].apply(preprocess)
df3['listed_in'] = df3['listed_in'].apply(preprocess)
df3['rating'] = df3['rating'].apply(preprocess)
```

```python
df3.sample(5)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')
nltk.download('punkt')

def data_preprocessing(text):
  #lower casing
  text = text.lower()

  #remove punctuations
  text = re.sub(r'[^\w\s]', '', text)

  #remove stopwords
  new_text = []
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear
  text =  ' '.join(x)

  #Tokenize
  text = word_tokenize(text)

  #Stemming
  new_list = []
  ps = PorterStemmer()
  for word in text:
    new_list.append(ps.stem(word))
  text = " ".join(new_list)

  return text
```

```python
data_preprocessing('Unleashing his inquisitive, intense comedic st')
```

```python
df3['description'] = df3['description'].apply(data_preprocessing)
```

```python
df3.sample(4)
```

```python
# List of columns whose values you want to concatenate
columns_to_concat = ['description','type', 'director', 'cast', 'country','rating','listed_in']

# Concatenate the values of specified columns into a single string separated by commas
df3['tag'] = df3[columns_to_concat].apply(lambda x: ','.join(x.dropna().astype(str)), axis=1)
```

```python
df3['tag'][4]
```

```python
df3.drop(columns_to_concat, axis=1, inplace=True)
```

```python
df3.head(3)
```

Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Create Tfidf vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features = 5000,binary=False)

#Fit and transform the array
vectors = tfidf_vectorizer.fit_transform(df3['tag']).toarray()
```

```python
vectors.shape
```

```python
tfidf_vectorizer.get_feature_names_out()[5:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
df.head()
```

```python
def recommend(title):
  title_index = df3[df3['title'] == title].index[0]
  distances = similarity[title_index]
  title_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
  llist = []
  for i in title_list:
    llist.append(i[0])
  for y in  llist:
    print(df3.iloc[y].title)

recommend('Kota Factory')
```
