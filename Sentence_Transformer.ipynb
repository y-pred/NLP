```python
from sentence_transformers import SentenceTransformer,util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.']

embeddings = model.encode(sentences)

for sentence,embedding in zip(sentences,embeddings):
  print(sentence)
  print(embedding)

```

Cosine Similarity

```python
emb1 = model.encode("I am eating a delicious Apple")
emb2 = model.encode("I am working on apple computer")
emb3 = model.encode('I am eating a fruit')
cos_sim = util.cos_sim(emb1,emb2)
print("Cosine Similarity:",cos_sim)
print("Cosine Similarity:",util.cos_sim(emb1,emb3))
```

Compute cosine similarity between all pairs

```python
sentences = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'The girl is carrying a baby.',
          'A man is riding a horse.',

          'Two men pushed carts through the woods.',
          'A man is riding white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in gorilla costume is playing a set of drums.'
          ]
embeddings = model.encode(sentences)

count=0
for sentence, embedding in zip(sentences,embeddings):
  print(sentence)
  print(embedding)
  count+=1
  if count ==2:
    break

```

```python
#compute cosine similarity
cos_sim = util.cos_sim(embeddings,embeddings)

print(cos_sim)
```

```python
#Print cosine combinations

cos_sim_combo = []
for i in range(len(cos_sim)-1):
  for j in range(i+1,len(cos_sim)):
    cos_sim_combo.append((cos_sim[i][j],i,j))
cos_sim_combo
```

```python
all_sentence_combinations=sorted(cos_sim_combo,key=lambda x: x[0],reverse = True)
```

```python
for score, i, j in all_sentence_combinations[0:5]:
    print(sentences[i],'______', sentences[j],'______', cos_sim[i][j])
```

Semantic Search

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('clips/mfaq')
```

```python
question = "<Q>How many hours do I need to work in office?"
answer_1 = "<A>According to the office HR policy, employee needs to work for 8 hours."
answer_2 = "<A>Not sure about this one"
answer_3 = "<A>Suresh mentioend he works for 10 hrs some times and some times 5 hours"

que_embedding = model.encode(question)
ans_embedding = model.encode([answer_1, answer_3, answer_3])
print(util.semantic_search(que_embedding,ans_embedding))
```

Clustering

```python
from sklearn.cluster import KMeans
import numpy as np

embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Corpus with example sentences
corpus = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'Horse is eating grass.',
          'A man is eating pasta.',
          'A Woman is eating Biryani.',
          'The girl is carrying a baby.',
          'The baby is carried by the woman',
          'A man is riding a horse.',
          'A man is riding a white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in a gorilla costume is playing a set of drums.',
          'A cheetah is running behind its prey.',
          'A cheetah chases prey on across a field.',
          'The cheetah is chasing a man who is riding the horse.',
          'man and women with their baby are watching cheetah in zoo'
          ]

corpus_embeddings = embedder.encode(corpus)
```

```python
corpus_embeddings[0]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
corpus_emb_scaled = scaler.fit_transform(corpus_embeddings)
corpus_emb_scaled
```

```python
clustering_model = KMeans(n_clusters=4)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```

```python
clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):

    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])

clustered_sentences
```
