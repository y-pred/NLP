```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd

```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
df = pd.read_csv('/content/Fake_Real_Data.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['label_encoded']=le.fit_transform(df['label'])
```

```python
df.head()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
stop_words = nlp.Defaults.stop_words
```

```python
def preprocess(text):
  doc = nlp(str(text))
  preprocessed_text = []
  for token in doc:
    if token.is_punct or token.like_num or token in stop_words or token.is_space:
      continue
    preprocessed_text.append(token.lemma_.lower().strip())
  return ' '.join(preprocessed_text)
```

```python
preprocess('   Hi! I am ashish. I love eating bananas. Would you go running with me?')
```

```python
df['preprocessed_text'] = df['Text'].apply(preprocess)
```

```python
df.sample(5)
```

```python
df['embedding'] = df['preprocessed_text'].apply(model.encode)
```

```python
df.head()
```

```python
X = df['embedding'].to_list()
y =df['label_encoded'].to_list()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
from sklearn import metrics
predicted = LR.predict(X_test)
print("Logistic Regression Accuracy:",metrics.accuracy_score(y_test, predicted))
print("Logistic Regression Precision:",metrics.precision_score(y_test, predicted,average='weighted'))
print("Logistic Regression Recall:",metrics.recall_score(y_test, predicted,average='weighted'))
```
